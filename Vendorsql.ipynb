Vendor/Sales Analysis with sql

```python
#En cox satis eden vendorlar
query="""
Select VendorNo,VendorName,sum(SalesDollars) from sales
group by VendorNo,VendorName
order by sum(SalesDollars) desc
limit 5
"""
pd.read_sql(query,conn)
```

```python
# Hansi vendorların satisi yoxdur?
query = """
SELECT v.VendorNumber, v.VendorName
FROM vendors v
LEFT JOIN sales s
ON v.VendorNumber = s.VendorNo
WHERE s.VendorNo IS NULL
"""

pd.read_sql(query, conn)
```

```python
#Hansi vendorlarin umumi satis meblegi en yuksekdir
query="""
SELECT VendorName, TotalSalesDollars
FROM vendor_sales
ORDER BY TotalSalesDollars DESC
LIMIT 5;
"""
pd.read_sql(query,conn)
```

```python
#Hər vendorun satış payını ümumi satışa görə hesablayaq.
query="""
SELECT  vs.VendorName,vs.TotalSalesDollars,vs.TotalSalesDollars * 1.0 / SUM(vs.TotalSalesDollars) OVER () AS Sales_Share
FROM vendor_sales vs
ORDER BY Sales_Share DESC
LIMIT 10;
"""
pd.read_sql(query,conn)
```

```python
# Vendorları performance a görə kateqoriyalayaq
#bunun ucun en yuksek satisimiz,ve en az satisimizi oyrenek
query="""
Select max(SalesDollars),min(SalesDollars) from sales
"""
pd.read_sql(query,conn)
#gorunduyu kimi bu 5000000 dan coxdur
#indi ise min baxaq
```

```python
# Vendorları performance a görə kateqoriyalayaq
query="""
SELECT VendorNo,Performance,COUNT(*) AS SalesCount
FROM (
    SELECT VendorNo,
        CASE
            WHEN SalesDollars <= 1000 THEN 'Low'
            WHEN SalesDollars BETWEEN 1000 AND 10000 THEN 'Medium'
            ELSE 'High'
        END AS Performance
    FROM sales
) t
where Performance=="Medium"
GROUP BY VendorNo, Performance
"""
pd.read_sql(query,conn)

```

```python
#Net Sales hesablayaq
query="""
Select VendorNumber,Vendorname,sum(TotalSalesDollars)-sum(TotalExcisetax) as Net_Sales
from vendor_sales
group by VendorNumber,VendorName
order by Net_Sales desc
"""
pd.read_sql(query,conn)
```

```python
#tranzaksiyalarin sayini tapaq her vendor ucun
query="""
SELECT VendorNumber,VendorName,COUNT(*) AS Number_of_Transactions
FROM vendor_sales
GROUP BY VendorNumber, VendorName
ORDER BY Number_of_Transactions DESC;
"""
pd.read_sql(query,conn)
#martignetti Companies tranzaksiya sayi coxdur
```

```python
#satis trendine baxaq
query="""
SELECT strftime('%Y-%m', InvoiceDate) AS month,SUM(TotalSalesDollars) AS total_sales
FROM vendor_sales
join vendor_invoice using(VendorNumber)
GROUP BY strftime('%Y-%m', InvoiceDate)
ORDER BY month;
"""
pd.read_sql(query,conn)
```

```python
#hansi Mehsullar daha cox satilib
query="""
SELECT InventoryId,SUM(SalesQuantity) AS Total_Quantity
FROM sales
GROUP BY InventoryId
ORDER BY Total_Quantity DESC
"""
pd.read_sql(query, conn)
```

```python
#Top 20% məhsullar ümumi satışın neçə %-ni verir?
query="""
WITH ranked AS (
    SELECT InventoryId,SUM(SalesQuantity) AS Total_Quantity,RANK() OVER (ORDER BY SUM(SalesQuantity) DESC) AS rank,COUNT(*) OVER () AS total_products,
        SUM(SUM(SalesQuantity)) OVER (ORDER BY SUM(SalesQuantity) DESC) * 1.0 /
        SUM(SUM(SalesQuantity)) OVER () AS cumulative_pct #Top mehsullardan baslayaraq satislarimizi yigib onu umumi satisa boluruk
      FROM sales
    GROUP BY InventoryId
)
SELECT ROUND(MAX(cumulative_pct) * 100, 2) AS top_20_percent_sales
FROM ranked
WHERE rank <= total_products * 0.2;
"""
pd.read_sql(query,conn)
# balansli bir satis neticesi
```

 Preto analizi ile bu analiz göstərir ki, satışlar müəyyən qədər konsentrasiyalıdır, lakin tam dominant deyil. Yəni həm əsas məhsullar, həm də digər məhsullar birlikdə ümumi gəlirin formalaşmasında rol oynayır.

```python
#StockTurnover(stokun neçə dəfə satılıb yeniləndiyi) yuksek olan mehsullari cixaraq
query="""
SELECT Description,AVG(StockTurnover) AS Avg_StockTurnover
FROM vendor_sales
GROUP BY Description
ORDER BY Avg_StockTurnover DESC
LIMIT 5
"""
pd.read_sql(query,conn)
```

```python
#Max
query="""
SELECT Description,MAX(StockTurnover) AS Max_StockTurnover
FROM vendor_sales
GROUP BY Description
ORDER BY Max_StockTurnover DESC
LIMIT 5
"""
pd.read_sql(query, conn)
#Buradan bunu anlayiriq ki məhsul bəzi dövrlərdə çox sürətli satılıb, amma stabil deyil
```

```python
#Stokda olubda hec satilmayan mehsulum yoxdur
query="""
SELECT
    Description,
    StockTurnover
FROM vendor_sales
WHERE StockTurnover = 0
ORDER BY StockTurnover
"""
pd.read_sql(query,conn)
```

```python
#Hansi Vendorlar daha cox alib ama az satib
query="""
Select VendorNumber,VendorName,TotalSalesDollars,TotalPurchaseDollars
from vendor_sales
where TotalSalesDollars<TotalPurchaseDollars
group by VendorNumber
order by TotalPurchaseDollars desc
"""
pd.read_sql(query,conn)
```

```python
#SalesToPurchaseRatio-(Satış/Alış nisbəti) rank edek Vendorlari ona gore
query="""
SELECT VendorNumber,VendorName,AVG(SalesToPurchaseRatio) AS Avg_Ratio,RANK() OVER (ORDER BY AVG(SalesToPurchaseRatio) DESC) AS Vendor_Rank
FROM vendor_sales
GROUP BY VendorNumber, VendorName
ORDER BY Avg_Ratio DESC;
"""
pd.read_sql(query,conn)
#AVG_Ratio 86.73 olanlarda satislar alislardan coxdur
```

```python
#her Vendor ucun ortalama ödəniş gecikməsi gunlerle
import pandas as pd
import sqlite3
query="""
SELECT v.VendorNumber,v.VendorName,round(AVG(JULIANDAY(v.PayDate) - JULIANDAY(v.InvoiceDate)),0) AS Avg_Payment_Delay
FROM vendor_invoice v
JOIN purchases p USING (VendorNumber)
GROUP BY v.VendorNumber, v.VendorName
ORDER BY Avg_Payment_Delay DESC;
"""
pd.read_sql(query,conn)
```
